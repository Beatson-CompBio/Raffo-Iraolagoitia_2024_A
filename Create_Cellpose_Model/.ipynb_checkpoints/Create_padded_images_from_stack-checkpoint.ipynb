# Create padded images from stack

```python
import tifffile as tf 
import numpy as np
import os
```

```python
image_folder = 'N:/SHARED/1 Week/Ryan Corbyn/FromGemma/training_image_cellpose/Stacks/'
image = 'C3-31072023_image5.tif'

image_data = np.array( tf.imread(image_folder + image) )
print(image_data.shape)
```

```python
def pad_with(vector, pad_width, iaxis, kwargs):

    pad_value = kwargs.get('padder', 0)

    vector[:pad_width[0]] = pad_value

    vector[-pad_width[1]:] = pad_value
```

```python
padded_image = []
for i in range(image_data.shape[0]): 
    pad_image = np.pad(image_data[0, ...], int(image_data.shape[1]*0.1), pad_with)
    padded_image.append(pad_image)
    
padded_image = np.array(padded_image)
print(padded_image.shape)
```

```python
import matplotlib.pyplot as plt

plt.imshow(padded_image[0,...])
```

```python
for i in range(padded_image.shape[0]):
    tf.imwrite(os.path.dirname(os.path.dirname(image_folder)) + '/Training_images/' 
               + image[0:-4] + '_subimage_' + str(i) +'.tif', padded_image[i, ...])
```
